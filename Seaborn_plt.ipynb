```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import seaborn as sns
from matplotlib.cm import get_cmap
```

```python
ncCountyResults = pd.read_csv('ncCountyResultsForMap.csv')
ncCountyResults.head()
```

```python
# Set seaborn style
sns.set(style="whitegrid")
plt.figure(figsize=(20, 10))
sns.barplot(x='county', y='total_reg_voters', data= ncCountyResults, palette='colorblind')
plt.xlabel('County')
plt.ylabel('Total Registered Voters')
plt.title('Total Registered Voters by County')
plt.xticks(rotation=90)
plt.show()
```
